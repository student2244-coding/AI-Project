# **Data Cleaning:**

```python
print(df['Item Fat Content'].unique())
```

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({
    'low fat': 'Low Fat',
    'LF': 'Low Fat',
    'reg': 'Regular'
})
```

```python
print(df['Item Fat Content'].unique())
```

The 'Total' entry in the 'Item Fat Content' column seems to be an anomalous entry, potentially a summary row, as it also has `NaN` for other columns. This row should be removed for cleaner data analysis.

```python
df = df[df['Item Fat Content'] != 'Total']
print(df['Item Fat Content'].unique())
```

Business Requirements

KPI's Requirements


```python
#Total Sales
total_sales = df['Sales'].sum()

#Average Sales
avg_sales = df['Sales'].mean()

#No of Items Sold
no_of_items_sold = df['Sales'].count()

#Average Ratings
avg_ratings = df['Rating'].mean()

#Display
print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.0f}")
print(f"No of Items Sold: {no_of_items_sold:,.0f}")
print(f"Average Rating: {avg_ratings:,.1f}")
```

# **Chart Requirement**

# **Total Sales by Fat Content**





```python
sales_by_fat_content = df.groupby('Item Fat Content')['Sales'].sum()

plt.figure(figsize=(8, 8)) # Explicitly set a larger figure size
plt.pie(sales_by_fat_content, labels=sales_by_fat_content.index, autopct='%.1f%%', startangle=90, textprops={'fontsize': 12})
plt.title('Sales by Fat Content', fontsize=14)
plt.axis('equal')
plt.show()
```

**Total sales by item Type**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by Item Type and calculate total sales
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

# Create figure
plt.figure(figsize=(10, 6))

# Create bar chart
bars = plt.bar(sales_by_type.index, sales_by_type.values)

# Rotate x-axis labels
plt.xticks(rotation=90)

# Labels and title
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

# Add value labels on top of bars
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():.0f}',
        ha='center',
        va='bottom',
        fontsize=8
    )

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by Outlet Establishment Year and calculate total sales
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_values(ascending=False)

# Create figure
plt.figure(figsize=(10, 6))

# Create bar chart
bars = plt.bar(sales_by_year.index.astype(str), sales_by_year.values)

# Rotate x-axis labels
plt.xticks(rotation=90)

# Labels and title
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment Year') # Updated title for clarity

# Add value labels on top of bars
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():.0f}',
        ha='center',
        va='bottom',
        fontsize=8
    )

plt.tight_layout()
plt.show()
```

**Fat Content by Outlet for Total Sales **

```python
import pandas as pd
import matplotlib.pyplot as plt

# If needed, load dataset
# df = pd.read_csv("BlinkIT.csv")

# Group by Outlet Location Type and Item Fat Content
grouped = df.groupby(
    ['Outlet Location Type', 'Item Fat Content']
)['Sales'].sum().unstack()

# Optional: Arrange columns in specific order
grouped = grouped[['Regular', 'Low Fat']]

# Plot bar chart
ax = grouped.plot(
    kind='bar',
    figsize=(8, 5),
    color=['orange', 'blue']
)

# Labels and title
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.title('Outlet Tier by Item Fat Content')

# Add value labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', fontsize=8)

plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

Total Sales by Outlet Establishment

```python
import pandas as pd
import matplotlib.pyplot as plt

# If needed, load dataset
# df = pd.read_csv("BlinkIT.csv")

# Group by Outlet Establishment Year and sum Sales
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

# Create figure
plt.figure(figsize=(9, 5))

# Plot line chart
plt.plot(
    sales_by_year.index,
    sales_by_year.values,
    marker='o',
    linestyle='-',
    color='blue'
)

# Labels and title
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment Year')

# Add value labels on points
for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

**Sales by Outlet Size**

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()
plt.figure(figsize=(4, 4))
plt.pie(sales_by_size, labels=sales_by_size.index, autopct='%1.1f%%', startangle=90)
plt.title('Outlet Size')
plt.tight_layout()
plt.show()
```

**Sales by Outlet Location**

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

plt.figure(figsize=(8,3))
ax = sns.barplot(x='Sales',y='Outlet Location Type', data=sales_by_location)

plt.title('Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

plt.tight_layout()
plt.show()
```
